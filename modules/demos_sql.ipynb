```python
from IPython.display import display
from SQLBuilder import SQLBuilder

# From raw
sql = SQLBuilder('thermo30.txt')
dfs = sql.sql2pandas('thermo30.sqlite')
for df in dfs:
    display(df)

# Save xml
sql = SQLBuilder('thermo30.txt', save_xml=True)

# From xml
sql = SQLBuilder('thermo30.xml')
dfs = sql.sql2pandas('thermo30.sqlite')
for df in dfs:
    display(df)
```
